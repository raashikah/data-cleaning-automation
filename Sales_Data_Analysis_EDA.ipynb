This notebook performs exploratory data analysis (EDA) on cleaned sales data.
The goal is to understand sales distribution, trends, and key drivers
to support business decision-making.


```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_sales_data.csv')  # change name if needed
df.head()
```

```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()
total_quantity = df['quantity'].sum()

total_sales, total_profit, total_quantity
```

```python
category_summary = df.groupby('category')[['sales', 'profit']].sum().sort_values(by='sales', ascending=False)
category_summary
```

```python
top_products = (
    df.groupby('product.name')['sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_products
```

```python
region_sales = df.groupby('region')['sales'].sum().sort_values(ascending=False)
region_sales
```

```python
sales_by_year = df.groupby('year')['sales'].sum()
sales_by_year
```

```python
import matplotlib.pyplot as plt

category_summary['sales'].plot(kind='bar', figsize=(8,4), title='Sales by Category')
plt.ylabel('Sales')
plt.show()
```

```python
sales_by_year.plot(kind='line', marker='o', title='Sales Trend Over Years')
plt.ylabel('Sales')
plt.xlabel('Year')
plt.show()
```

Key Insights:
1. Technology category contributes the highest sales and profit.
2. A small set of products drives a large share of revenue.
3. Sales show an increasing trend over the years, indicating business growth.
4. Certain regions contribute high sales but relatively lower profit, suggesting pricing or logistics optimization opportunities.
